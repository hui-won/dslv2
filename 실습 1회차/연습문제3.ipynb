```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsRegressor
```

```python
df=pd.read_csv("set_03_data.csv")
```

```python
df_q1=df.copy()
df_q1
```

```python
sales_mean=df_q1["sales_idx"].mean()
sales_std=df_q1["sales_idx"].std()
standard=sales_mean+2*sales_std
standard
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df_q2=df.copy()
df_q2=df_q2.drop(columns="screen_size")
df_q2
```

```python
ss=StandardScaler().fit(df_q2)
df_q2=pd.DataFrame(ss.transform(df_q2),columns=df_q2.columns)
df_q2
```

```python
model_km=KMeans(n_clusters=4, random_state=123)
out_km=model_km.fit(df_q2)
out_km.cluster_centers_
df_out_km=pd.DataFrame(out_km.cluster_centers_,columns=df_q2.columns)
df_out_km
```

```python
from sklearn.metrics import euclidean_distances
```

```python
euclidean_distances(df_out_km)
```

```python
2.66
```

```python
df_q3=df.copy()
df_q3
```

```python
df_q3=pd.get_dummies(df_q3,columns=["screen_size"])
df_q3=df_q3.set_index("sales_idx").reset_index()
df_q3
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x=train_test_split(df_q3,test_size=0.2,random_state=123)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mmScaler=MinMaxScaler().fit(train_x)
arr_train_x=mmScaler.transform(train_x)
arr_test_x=mmScaler.transform(test_x)
arr_train_x.shape, arr_test_x.shape

df_train_x=pd.DataFrame(arr_train_x,columns=train_x.columns)
df_test_x=pd.DataFrame(arr_test_x,columns=test_x.columns)
```

```python
df_train_x.info()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
k_list=[3,5,7,9,11]

for k in k_list:
    model_knn=KNeighborsRegressor(n_neighbors=k)
    out_knn=model_knn.fit(X=df_train_x.iloc[:,1:],y=df_train_x["sales_idx"])
    pred=out_knn.predict(X=df_test_x.iloc[:,1:])

    print(mean_squared_error(y_true=df_test_x["sales_idx"],y_pred=pred)**0.5)
    
```

```python
model_best=KNeighborsRegressor(n_neighbors=3)
out_knn_best=model_best.fit(X=df_train_x.iloc[:,1:],y=df_train_x["sales_idx"])
```
