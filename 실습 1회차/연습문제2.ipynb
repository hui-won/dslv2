```python
df_q3=df.copy()
df_q3
```

```python
df_q3_x=pd.DataFrame()
```

```python
df_q3["Churn"].unique()
```

```python
df_q3_x["SeniorCitizen"]=df_q3["SeniorCitizen"]
df_q3_x["Partner"]=np.where(df_q3["Partner"]=="Yes",1,0)
df_q3_x["Dependents"]=np.where(df_q3["Dependents"]=="Yes",1,0)
df_q3_x["tenure"]=df_q3["tenure"]
df_q3_x["MonthlyCharges"]=df_q3["MonthlyCharges"]
df_q3_x["TotalCharges"]=df_q3["TotalCharges"]
df_q3_x["OnlineSecurity"]=df_q3["OnlineSecurity"].replace({"Yes":1,"No":0})
df_q3_x["OnlineSecurity"]=np.where((df_q3_x["OnlineSecurity"]!=0)&(df_q3_x["OnlineSecurity"]!=1),-1,df_q3_x["OnlineSecurity"])
df_q3_x["OnlineBackup"]=df_q3["OnlineBackup"].replace({"Yes":1,"No":0})
df_q3_x["OnlineBackup"]=np.where((df_q3_x["OnlineBackup"]!=0)&(df_q3_x["OnlineBackup"]!=1),-1,df_q3_x["OnlineBackup"])
df_q3_x["DeviceProtection"]=df_q3["DeviceProtection"].replace({"Yes":1,"No":0})
df_q3_x["DeviceProtection"]=np.where((df_q3_x["DeviceProtection"]!=0)&(df_q3_x["DeviceProtection"]!=1),-1,df_q3_x["DeviceProtection"])
df_q3_x["TechSupport"]=df_q3["TechSupport"].replace({"Yes":1,"No":0})
df_q3_x["TechSupport"]=np.where((df_q3_x["TechSupport"]!=0)&(df_q3_x["TechSupport"]!=1),-1,df_q3_x["TechSupport"])
df_q3_x["StreamingMovies"]=df_q3["StreamingMovies"].replace({"Yes":1,"No":0})
df_q3_x["StreamingMovies"]=np.where((df_q3_x["StreamingMovies"]!=0)&(df_q3_x["StreamingMovies"]!=1),-1,df_q3_x["StreamingMovies"])
df_q3_x["PaperlessBilling"]=df_q3["PaperlessBilling"].replace({"Yes":1,"No":0})
df_q3_x
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df_q3_x["Churn"]=df_q3["Churn"].replace({"Yes":1,"No":0})
```

```python
df_q3_x["customerID"]=df_q3["customerID"].str[0:4].astype("int")
train_x=df_q3_x.loc[df_q3_x["customerID"]%4!=0]
test_x=df_q3_x.loc[df_q3_x["customerID"]%4==0]
train_x=train_x.drop(columns="customerID")
test_x=test_x.drop(columns="customerID")
train_x.shape, test_x.shape
```

```python
train_xx=train_x.iloc[:,:-1]
test_xx=test_x.iloc[:,:-1]
```

```python
mmScaler = MinMaxScaler().fit(train_xx)
train_xx=pd.DataFrame(mmScaler.transform(train_xx),columns=train_xx.columns)
test_xx=pd.DataFrame(mmScaler.transform(test_xx),columns=test_xx.columns)
```

```python
model_lr=LogisticRegression(random_state=123)
out_lr=model_lr.fit(X=train_xx,y=train_x["Churn"])

pred=out_lr.predict(X=test_xx)

f1_score(y_true=test_x["Churn"], y_pred=pred)
```
