```python
df_q2=df.copy()
df_q2.info()
```

```python
df_q2=df.loc[(df_q2["Operating_system"].str.contains("4K"))|(df_q2["channel"].str.contains("4K"))|(df_q2["Picture_quality"].str.contains("4K"))]
df_q2
```

```python
condi_1=df_q2["Stars"]==0
condi_2=df_q2["Ratings"]==0
condi_3=df_q2["Reviews"]==0

df_q2=df_q2.loc[~(condi_1|condi_2|condi_3)]
df_q2.info()
```

```python
df_q2_n=df_q2.iloc[:,1:6]
df_q2_n
```

```python
model_km=KMeans(n_clusters=3, random_state=123)
out_km=model_km.fit(df_q2_n)
pd.Series(out_km.labels_).value_counts()
```

```python
df_q3=df.copy()
df_q3
```

```python
df_q3=df_q3.loc[~df_q3["channel"].str.contains("Pixel|Oper")]
df_q3.info()
```

```python
df_q3_x=pd.DataFrame()
df_q3_x["후기작성비율"]=df_q3["Reviews"]/df_q3["Ratings"]
df_q3_x["MRP"]=df_q3["MRP"]
df_q3_x["할인율"]=df_q3["current_price"]/df_q3["MRP"]
df_q3_x["Netflix"]=np.where(df_q3["channel"].str.contains("Netflix"),1,0)
df_q3_x["Prime"]=np.where(df_q3["channel"].str.contains("Prime Video"),1,0)
df_q3_x["고해상도"]=np.where(df_q3["Picture_quality"].str.contains("4K|8K"),1,0)
df_q3_x["Stars"]=df_q3["Stars"]
df_q3_x=df_q3_x.dropna()
df_q3_x
```

```python
model_rf=RandomForestRegressor(random_state=123)
out_rf=model_rf.fit(X=df_q3_x.drop(columns="Stars"),y=df_q3_x["Stars"])

pd.Series(out_rf.feature_importances_).value_counts()
```
